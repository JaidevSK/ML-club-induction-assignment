Defining the kernel functions in the following cell as well as defining the kernel matrix and the mean predicted value of y. Here we can add the values of the sigmaf, and the l as required as they were defined in the attached in the handwritten text. Here, the value of sigmaf and l were substituted according to the appropriate reasons as mentioned in the text.

```python
from math import exp
import numpy as np
def k(x, X, sigmaf=0.1, l=20):
  return (((sigmaf)**2)*exp(-((x-X)**2)/(2*l)))
def gpr(x1):
  K=np.array([[k(1,1), k(1,2), k(1,3)], [k(2,1), k(2,2), k(2,3)], [k(3,1), k(3,2), k(3,3)]])
  K1=np.array([k(x1, 1), k(x1, 2), k(x1, 3)])
  K2=np.array([k(x1, x1)])
  y1bar=np.matmul(np.matmul(K1,np.linalg.inv(K)), y)
  return y1bar
```

Plotting the Actual, Predicted and Extrapolated points to visualise the working of the model

```python
import matplotlib.pyplot as plt

plt.scatter([1,2,3,4], [3,4,5,6], c='r', label="Actual Y")
plt.plot([1,2,3,4], [gpr(1),gpr(2),gpr(3),gpr(4)], c='b', label="Predicted Y")
plt.legend()
```
